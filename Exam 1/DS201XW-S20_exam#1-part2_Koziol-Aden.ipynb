```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Automobile_price_data_clean-f18.csv')

# Question 1
print("Question 1:")
average = df.groupby('make')['price'].mean().reset_index()
plt.bar(average['make'], average['price'])
plt.xticks(rotation=90, ha='center')
plt.show()

# Question 2
print("Question 2:")
above8000 = df[(df['make'] == 'toyota') & (df['price'] > 8000)]
numAbove = len(above8000)
print(str(numAbove) + " toyota cars priced above $8,000")

# Question 3
print("Question 3:")
hatchback = df[df['body-style'] == 'hatchback']
avgMpg = hatchback['city-mpg'].mean()
print("Average city mpg for a hatchback: " + str(avgMpg))

# Question 4
print("Question 4:")
plt.figure(figsize=(15, 10))
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.boxplot([df[df['body-style'] == style]['price'].dropna().values for style in df['body-style'].unique()], labels=df['body-style'].unique())
plt.show()

# Question 5
print("Question 5:")
plt.figure(figsize=(15, 10))
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.boxplot([df[df['make'] == make]['price'].dropna().values for make in df['make'].unique()], labels=df['make'].unique())
plt.show()
```
