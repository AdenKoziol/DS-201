```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("USvideos.csv")

# Explore the Data
print(df.head())
print(df.info())
df.drop_duplicates(inplace=True)

# Work on your dataset
# Step 1
correlation_matrix = df[['views', 'likes', 'category_id']].corr()
# Step 3
top_100_views = df.nlargest(100, "views")
# Step 4
top_10_likes_views = df.nlargest(10, ['likes','views'])
# Step 5
df_top100views = df.nlargest(100, "views")

# Create the following charts
# Chart 1
sns.pairplot(df[["category_id", "views", "likes"]], palette='deep', diag_kind="hist")
plt.show()
# Chart 2
sns.pairplot(df_top100views[["category_id", "views", "likes"]], palette='deep', diag_kind="hist")
plt.show()

# Observe and describe the differences from those two plots
# The first pairplot shows the relationships between views, likes, and category_id for all videos in the dataset.
# The second pairplot shows the same relationships but only for the top 100 videos by views. 
# Differences include a lot more spread out data along the diagonal of the second chart.
```
