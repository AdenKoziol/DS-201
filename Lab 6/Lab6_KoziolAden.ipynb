```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#step 2
df = pd.read_csv('http://bit.ly/drinksbycountry') 

#step 3
europedf = df[df['continent'] == 'Europe']
beerEurope = europedf[europedf['beer_servings'] > 50]
print(beerEurope[['country', 'beer_servings']])

#step 4
continent_means = df.drop('country', axis=1).groupby('continent').mean()
colors = ['red', 'green', 'blue', 'orange']
ax = continent_means.plot(kind='bar', figsize=(15, 7), color=colors, width=0.8)
plt.xlabel('Continent')
plt.legend(loc='upper left')
plt.show()

#step 5
beer_by_continent = df.groupby('continent')['beer_servings'].mean()
beer_by_continent.plot(kind='bar', color='blue', figsize=(10, 6))
plt.title('Bar graph grouped by continent')
plt.xlabel('Continent')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='continent', y='beer_servings', data=df)
plt.title('Boxplot grouoped by continent')
plt.xlabel('Continent')
plt.show()
```
