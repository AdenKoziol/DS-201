```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('lab5.csv') 
df.head()
df.shape

df2 = df.groupby(['State_name']).size()
type(df2)
colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan', 'magenta', 'yellow', 'brown', 'gray']

ax = df2.plot.bar(figsize=(15,7), color=colors)
ax.set_ylabel('# of Schools')
ax.set_title('# of Schools per State')
plt.show()
```
