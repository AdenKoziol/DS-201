```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/asukul/DS201/master/datasets/Lemonade2016-2.csv')

# Cleanse Data
df['Date'] = pd.to_datetime(df['Date'])
df.interpolate(inplace=True)
df.drop_duplicates(inplace=True)

# Add Columns
df['Sales'] = df['Lemon'] + df['Orange']
df['Revenue'] = df['Sales'] * df['Price']

# Chart 1
sns.scatterplot(x='Leaflets', y='Sales', data=df)
plt.show()

# Chart 2
sns.histplot(df['Revenue'], bins=10)
plt.show()
print("Graph is right skewed")

# Chart 3
sns.boxplot(x='Location', y='Sales', data=df)
plt.show()

# Chart 4
sns.boxplot(x='Location', y='Sales', data=df[df['Temperature'] >= 75])
plt.show()

# Chart 5
columnsToPlot = ['Revenue', 'Temperature', 'Leaflets', 'Location']
sns.pairplot(df[columnsToPlot], hue='Location', diag_kind='hist')
plt.show()
```
